## Significant wave of 0.5m to 2.5m

```python
wave_heights = np.arange(0.5, 2.51, 0.5)
for wave_height in wave_heights:
    !../build/asv_simulator ./asv $wave_height 0.0 1
```

```python
data = pd.read_csv('./asv_out', sep=" ")
```

```python
data.head()
```

```python
data.info()
```

### Path

```python
dfs = DataFrame(data, columns=['sig_wave_ht(m)', 'cog_x(m)', 'cog_y(m)']).groupby('sig_wave_ht(m)')
for (wave_ht, df) in dfs:
    plt.plot('cog_x(m)', 'cog_y(m)', data=df, label=wave_ht)
    plt.legend()
    plt.xlabel('cog_x(m)')
    plt.ylabel('cog_y(m)')
```

### Relative heave

```python
df = DataFrame(data, columns=['sig_wave_ht(m)', 'time(sec)', 'wave_elevation(m)', 'cog_z(m)'])
df['rel_heave(m)'] = df['cog_z(m)'] - df['wave_elevation(m)']
df['rel_heave(m)'] = df['rel_heave(m)'].abs()

g = sns.violinplot('sig_wave_ht(m)', 'rel_heave(m)', hue='sig_wave_ht(m)', data=df)
# Put a legend to the right side
g.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

### Pitch

```python
df = DataFrame(data, columns=['sig_wave_ht(m)', 'trim(deg)'])
df['abs_trim(deg)'] = df['trim(deg)'].abs()
g = sns.violinplot('sig_wave_ht(m)', 'abs_trim(deg)', hue='sig_wave_ht(m)', data=df)
# Put a legend to the right side
g.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

### Roll

```python
df = DataFrame(data, columns=['sig_wave_ht(m)', 'heel(deg)'])
df['abs_roll(deg)'] = df['heel(deg)'].abs()
g = sns.violinplot('sig_wave_ht(m)', 'abs_roll(deg)', hue='sig_wave_ht(m)', data=df)
# Put a legend to the right side
g.legend(loc='center right', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

### Surge velocity and acceleration

```python
dfs = DataFrame(data, columns=['sig_wave_ht(m)', 'time(sec)', 'surge_vel(m/s)', 'surge_acc(m/s2)']).groupby('sig_wave_ht(m)')
for (wave_ht, df) in dfs:
    plt.plot('time(sec)', 'surge_vel(m/s)', data=df, label=wave_ht)
    plt.plot('time(sec)', 'surge_acc(m/s2)', data=df, label=wave_ht)
    plt.legend()
    plt.xlabel('time(sec)')
    plt.ylabel('vel/acc)')
```

```python
# End of this section of test. Rename output file so that the next section does not overwrite it. 
! mv -v ./asv_out ./asv_out_wave_heights
```
